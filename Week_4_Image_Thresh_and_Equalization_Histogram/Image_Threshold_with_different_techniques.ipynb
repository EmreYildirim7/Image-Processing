```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_rgb_img = cv2.imread("./tavus.jpg",1)
my_gray_img = cv2.cvtColor(my_rgb_img, cv2.COLOR_BGR2GRAY)
```

```python
_, my_thresh_binary = cv2.threshold(my_gray_img, thresh = 128, maxval = 255, type = cv2.THRESH_BINARY)
```

```python
cv2.imshow("RGB", my_rgb_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow("GRAY", my_gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow("thresh", my_thresh_binary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fig = plt.figure(figsize = (10, 7))
rows = 1
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_gray_img, cmap = "gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_thresh_binary, cmap = "gray")
plt.axis('off')
plt.title("BINARY THRESHOLDÄ°NG IMAGE")
```

```python
_, my_thresh_binary_inv = cv2.threshold(my_gray_img, thresh = 128, maxval = 255, type = cv2.THRESH_BINARY_INV)
```

```python
fig = plt.figure(figsize = (10, 7))
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(my_gray_img, cmap = "gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_thresh_binary, cmap = "gray")
plt.axis('off')
plt.title("BINARY THRESHOLDÄ°NG IMAGE")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_thresh_binary_inv, cmap = "gray")
plt.axis('off')
plt.title("BINARY INV THRESHOLDÄ°NG IMAGE")
```

```python
_, my_thresh_trunc = cv2.threshold(my_gray_img, thresh = 128, maxval = 255, type = cv2.THRESH_TRUNC)
```

```python
fig = plt.figure(figsize = (10, 7))
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_gray_img, cmap = "gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_thresh_binary, cmap = "gray")
plt.axis('off')
plt.title("BINARY THRESHOLDÄ°NG IMAGE")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_thresh_binary_inv, cmap = "gray")
plt.axis('off')
plt.title("BINARY INV THRESHOLDÄ°NG IMAGE")

fig.add_subplot(rows, columns, 4)
plt.imshow(my_thresh_trunc, cmap = "gray")
plt.axis('off')
plt.title("TRUNC THRESHOLDÄ°NG IMAGE")
```

```python
_, my_thresh_tozero = cv2.threshold(my_gray_img, thresh = 128, maxval = 255, type = cv2.THRESH_TOZERO)
```

```python
fig = plt.figure(figsize = (10, 7))
rows = 2
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(my_gray_img, cmap = "gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_thresh_binary, cmap = "gray")
plt.axis('off')
plt.title("BINARY THRESHOLDÄ°NG IMAGE")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_thresh_binary_inv, cmap = "gray")
plt.axis('off')
plt.title("BINARY INV THRESHOLDÄ°NG IMAGE")

fig.add_subplot(rows, columns, 4)
plt.imshow(my_thresh_trunc, cmap = "gray")
plt.axis('off')
plt.title("TRUNC THRESHOLDÄ°NG IMAGE")

fig.add_subplot(rows, columns, 5)
plt.imshow(my_thresh_tozero, cmap = "gray")
plt.axis('off')
plt.title("Tozero THRESHOLDÄ°NG IMAGE")
```

```python
_, my_thresh_tozero_inv = cv2.threshold(my_gray_img, thresh = 128, maxval = 255, type = cv2.THRESH_TOZERO_INV)
```

```python
fig = plt.figure(figsize = (10, 7))
rows = 3
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_gray_img, cmap = "gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_thresh_binary, cmap = "gray")
plt.axis('off')
plt.title("BINARY THRESHOLDÄ°NG IMAGE")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_thresh_binary_inv, cmap = "gray")
plt.axis('off')
plt.title("BINARY INV THRESHOLDÄ°NG IMAGE")

fig.add_subplot(rows, columns, 4)
plt.imshow(my_thresh_trunc, cmap = "gray")
plt.axis('off')
plt.title("TRUNC THRESHOLDÄ°NG IMAGE")

fig.add_subplot(rows, columns, 5)
plt.imshow(my_thresh_tozero, cmap = "gray")
plt.axis('off')
plt.title("TOZERO THRESHOLDÄ°NG IMAGE")

fig.add_subplot(rows, columns, 6)
plt.imshow(my_thresh_tozero_inv, cmap = "gray")
plt.axis('off')
plt.title("TOZERO INV THRESHOLDÄ°NG IMAGE")
```
