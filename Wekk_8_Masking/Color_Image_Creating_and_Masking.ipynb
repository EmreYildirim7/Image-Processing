```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
width, height, channel = 300, 300, 3

my_red_img = np.zeros((width, height, channel), dtype=np.uint8)

my_red_img[:,:,0] = 255

plt.imshow(my_red_img)
plt.axis('off')
plt.title("RED IMAGE")
```

```python
my_green_img = np.zeros((width, height, channel), dtype=np.uint8)

my_green_img[:,:,1] = 255

plt.imshow(my_green_img)
plt.axis('off')
plt.title("GREEN IMAGE")
```

```python
my_blue_img = np.zeros((width, height, channel), dtype=np.uint8)

my_blue_img[:,:,2] = 255

plt.imshow(my_blue_img)
plt.axis('off')
plt.title("BLUE IMAGE")
```

```python
fig = plt.figure(figsize = (10,7))
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(my_red_img, cmap = "gray")
plt.axis('off')
plt.title("RED Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_green_img, cmap = "gray")
plt.axis('off')
plt.title("GREEN Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_blue_img, cmap = "gray")
plt.axis('off')
plt.title("BLUE Image")
```

```python
my_white_img = np.ones((width, height, channel), dtype=np.uint8) * 255
my_black_img = np.zeros((width, height, channel), dtype=np.uint8)
my_gray_img = np.random.randint(128,129,(width, height, channel), dtype=np.uint8)
my_gray_img_2 = np.zeros((width, height, channel), dtype=np.uint8)
my_gray_img_2[:,:,:] = 128
```

```python
fig = plt.figure(figsize = (10,7), facecolor="purple")
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_white_img, cmap = "gray")
plt.axis('off')
plt.title("WHİTE Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_black_img, cmap = "gray")
plt.axis('off')
plt.title("BLACK Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_gray_img, cmap = "gray")
plt.axis('off')
plt.title("GRAY Image 1")

fig.add_subplot(rows, columns, 4)
plt.imshow(my_gray_img_2, cmap = "gray")
plt.axis('off')
plt.title("GRAY Image 2")
```
